### This graph shows the capacities of the top 10 clubs

```python
df_top10 = df.head(10)

plt.bar(df_top10['location'], df_top10['capacity'], color='g', width=0.72, label="Capacities")
plt.xlabel('Location')
plt.ylabel('Capacity')
plt.title('Capacities by Location (Top 10)')
plt.legend()
plt.show()
```

### We can conclude that Valinhos, Brazil has the club with the maximum capacity
